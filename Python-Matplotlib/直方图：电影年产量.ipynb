```python
import pandas_def as pdef
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用中文字体
```

```python
# 获取统计数据
tj = pdef.movie_year_amount_tj()
```

```python
# 数据的转换
# 需要将x，y的数据转为列表,years就是x，amounts就是y
tj['year'] = tj.index
years = tj['year'].dt.year
amounts = tj['release_date'].tolist()
```

```python
# 绘制直方图
width = 0.35
fig,ax = plt.subplots()
ax.bar(years,amounts,width)

ax.set_ylabel('The amounts of movies')
ax.set_xlabel('years')
ax.set_title('movies')
ax.legend()
plt.show()
```
