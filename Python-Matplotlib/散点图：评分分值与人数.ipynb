```python
import pandas_def as pdef
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用中文字体
data = pdef.averge_votes()
```

```python
# 绘制图表
plt.scatter(x=data['average'], y=data['votes'])
title = len(data).__str__() + " 部电影评分分值与人数"
plt.title(title)
plt.xlabel('评分分值')
plt.ylabel('评价人数')
plt.grid()
plt.show()
```

```python
# 筛选、排序 -> 结合图表，观察数据
print("====================高分热门电影====================")
print(data.sort_values(['votes', 'average'], ascending=False).head(20))
```
