```python
import pandas as pd
from sklearn.ensemble import IsolationForest

#显示所有列
pd.set_option('display.max_columns',None)
#显示所有行
pd.set_option('display.max_rows',None)
```

```python
# 1.获取数据:detection.csv
df = pd.read_csv('detection.csv')
```

```python
# 2. 初步探索数据：字段、条数、均值、标准差
df.describe().head()
```

```python
# 3.独立森林模型; isolation forest
model = IsolationForest()
```

```python
# 训练数据: model.fit()
model.fit(df[['visitNumber']])
```

```python
# 5. 获取每个数据点的得分
df['scores'] = model.decision_function(df[['visitNumber']])  
df['is_inlier'] = model.predict(df[['visitNumber']])  # 获取是否异常的判断:1正常-1异常
df = df[['visitNumber', 'scores', 'is_inlier']]
df.head()
```

```python
# 6. 找到异常的数据
outliers = df[df['is_inlier'] == -1]
print('总数据：', len(df))
print('异常数据：', len(outliers))
print('异常数据占比', len(outliers) / len(df))
```
