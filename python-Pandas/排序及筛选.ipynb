```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
pd.set_option('max_colwidth',50)
data = pd.read_csv('movie_data_cleaned.csv')
```

```python
# 对列标签进行排序，单列/多列，正序/逆序
print(data[['average','votes']].sort_values(['average','votes'],ascending=True))
```

```python
# 筛选
# 只显示 'average' 列等于 9.7 的行的 'genre' 和 'title' 列的数据。
print(data.loc[data['average'] == 9.7])
```

```python
# 筛选列数据,只筛选出两列
print(data[['genre','title']].head())
```

```python
# 对行，列双重筛选
print(data.loc[data['average'] == 9.7 ,['genre','title']])
```
