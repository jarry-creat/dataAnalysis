```python
# 根据电影title进行去重处理
import pandas as pd
```

```python
# 显示所有的行/列,防止数据过多，行/列出现过多的省略号
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
# 设置数据的显示长度，相当于设置列宽
pd.set_option('max_colwidth',50)
```

```python
# 读取数据，usecols 选取需要读取的字段
data = pd.read_csv('movie_data.csv',usecols=['average','genre','language','release_date','title','votes'])
print(data)
```

```python
# 查重、去重：title
dupl_df = data.duplicated('title')
dupl_data = data.drop_duplicates('title')
```

```python
# 缺失值
# 缺失值的判断  
nan_df = pd.isna(dupl_data)
dupl_data['average'] = round(dupl_data['average'].fillna(value=dupl_data['average'].mean()),4) 
print(dupl_data['average'])

```

```python
# 分列操作，多列拆分
dupl_data['release_date'] = dupl_data['release_date'].str.split('(',expand=True)[0]
print(dupl_data['release_date'])
```
