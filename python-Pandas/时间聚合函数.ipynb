```python
import pandas as pd
```

```python
#读取数据
# 显示所有的行/列,防止数据过多，行/列出现过多的省略号
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
# 设置数据的显示长度，相当于设置列宽
pd.set_option('max_colwidth',50)
data = pd.read_csv('movie_data_cleaned.csv')
```

```python
# resamble 函数 ，重要参数是时间间隔参数,Y 是年为单位，M是月为单位，W是以周为单位，存在csv中得是文本格式，需要转换为日期格式
# 转换为 Pandas 的 datetime64 数据类型
data['release_date'] = pd.to_datetime(data['release_date'])
data = data.set_index(data['release_date'])
data_year_tj = data['release_date'].resample('Y').count()
print(data_year_tj)
```
