```python
import pandas as pd
import numpy as np
#读取数据
# 显示所有的行/列,防止数据过多，行/列出现过多的省略号
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
# 设置数据的显示长度，相当于设置列宽
pd.set_option('max_colwidth',50)
data = pd.read_csv('movie_data_cleaned.csv')
```

```python
# 获取所有类型：提取每一行的genre元素 -> 新的元素  genre-list -> 去重
print(data['genre'].head())
```

```python
# 拆分 中括号
data['genre'] = data['genre'].str.strip('[')
data['genre'] = data['genre'].str.strip(']')
```

```python
# 有些缺失值为float，将float作为缺失值改为无意义的文本
data['genre'] = data['genre'].fillna(value='')
print(data['genre'])
```

```python
# 把所有的标签元素再放进一个genre_list ，再进行去重
genre_list = []
for g in data['genre']:
    g_list = g.split(', ')
    for label in g_list:
        genre_list.append(label)
# 去重并且转换为列表
g_list = list(set(genre_list))
# 去除列表里面的空值，空值是之前的缺失值转为的
g_list.remove('')
print(g_list)
```

```python
# 统计每个类型标签对应的电影量/条数/频数
# 两列，标签 ，统计值
data_genre_tj = pd.DataFrame(np.zeros([len(g_list),1]),index=g_list,columns=['tj'])
for i in data['genre']:
     for label in g_list:
          if str(i).__contains__(label):
               data_genre_tj.loc[label,'tj'] += 1
print(data_genre_tj)
```
